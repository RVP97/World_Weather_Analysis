```python
import numpy as np
from citipy import citipy
import requests
# from config import weather_api_key
import time
import pandas as pd
import sys
sys.path.append('..')
from config import weather_api_key
```

```python
lat = np.random.uniform(low=-90, high=90, size=2000)
lng = np.random.uniform(low=-180, high=180, size=2000)
coordinates = zip(lat, lng)
```

```python
cities = []
for lat,lng in coordinates:
    city = citipy.nearest_city(lat, lng).city_name
    if city not in cities:
        cities.append(city)
```

```python
len(cities)
```

```python
base_url = 'https://api.openweathermap.org/data/2.5/weather'
set_count = 1
city_count = 1
city_data = []

for index, city in enumerate(cities):
    print(f'Processing record #{city_count} for set #{set_count}')
    if city_count % 50 == 0:
        set_count += 1
        time.sleep(30)
    city_count += 1
    parameters = {'q': city, 'appid': weather_api_key, 'units': 'imperial'}
    try:
        response = requests.get(base_url, params=parameters).json()
        country = response['sys']['country']
        lat = response['coord']['lat']
        lng = response['coord']['lon']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        description = response['weather'][0]['description']
        
        city_data.append({
            'City': city.title(),
            'Country': country,
            'Lat': lat,
            'Lng': lng,
            'Max Temp': max_temp,
            'Humidity': humidity,
            'Cloudiness': cloudiness,
            'Wind Speed': wind_speed,
            'Current Description': description
        })
    except:
        print(f'{city} no found')
print('________________________')
print('Data extraction Finished')
print('________________________')
```

```python
city_data_df = pd.DataFrame(city_data)
```
